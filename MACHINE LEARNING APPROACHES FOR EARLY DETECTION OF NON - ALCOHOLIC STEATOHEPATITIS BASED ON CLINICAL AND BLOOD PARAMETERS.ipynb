# Task
Develop a machine learning model using Logistic Regression to diagnose the stage of steatohepatitis. Train the model on 70% of the data and test it on 30%. Calculate and present the accuracy rate, and visualize the relationship between the predicted and actual steatosis stages. Use the data from "steatosis_hepatitis_data.csv".

## Data preparation

### Subtask:
Handle missing values and convert categorical features into numerical format.


**Reasoning**:
Check for missing values, fill them using the mean, and then one-hot encode the 'sex' column. Finally, display the head of the processed DataFrame.



```python
print("Missing values before handling:")
print(df.isnull().sum())

# Fill missing values with the mean
df.fillna(df.mean(numeric_only=True), inplace=True)

print("\nMissing values after handling:")
print(df.isnull().sum())

# Convert 'sex' column to numerical using one-hot encoding
df = pd.get_dummies(df, columns=['sex'], drop_first=True)

print("\nDataFrame after handling missing values and converting categorical features:")
display(df.head())
```

## Data splitting

### Subtask:
Split the data into training and testing sets (70% train, 30% test).


**Reasoning**:
Split the data into features (X) and the target variable (y), and then split these into training and testing sets.



```python
from sklearn.model_selection import train_test_split

X = df.drop('Steatosis stage', axis=1)
y = df['Steatosis stage']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

## Model training

### Subtask:
Train a Logistic Regression model on the training data.


**Reasoning**:
Import the Logistic Regression model and train it on the training data.



```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

**Reasoning**:
The error indicates that the target variable `y_train` has continuous values, but Logistic Regression is a classifier and expects discrete class labels. I need to convert the target variable to an integer type to represent discrete stages.



```python
y_train = y_train.astype(int)
y_test = y_test.astype(int)

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

## Model evaluation

### Subtask:
Evaluate the model's performance on the testing data and calculate the accuracy.


**Reasoning**:
Evaluate the model's performance on the testing data and calculate the accuracy.



```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f"The accuracy of the Logistic Regression model is: {accuracy:.4f}")
```

## Data visualization

### Subtask:
Visualize the relationship between the predicted and actual steatosis stages.


**Reasoning**:
Create a scatter plot to visualize the relationship between predicted and actual steatosis stages.



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.title("Predicted vs. Actual Steatosis Stage")
plt.xlabel("Actual Steatosis Stage")
plt.ylabel("Predicted Steatosis Stage")
plt.grid(True)
plt.show()
```

## Summary:

### Data Analysis Key Findings

*   Missing values were present in several columns of the dataset and were handled by imputing the mean for numerical columns and using one-hot encoding for the categorical 'sex' column.
*   The dataset was successfully split into training (70%) and testing (30%) sets, with 629 samples for training and 270 for testing.
*   A Logistic Regression model was trained to predict the steatosis stage. The target variable was converted to an integer type for model training.
*   The Logistic Regression model achieved an accuracy of 0.5370 on the test set.
*   A scatter plot visualized the relationship between the predicted and actual steatosis stages, showing some correlation but also indicating prediction errors.

### Insights or Next Steps

*   The current model accuracy of 53.7% is relatively low, suggesting that the model may not be highly effective at diagnosing the steatosis stage based on the provided features.
*   Further steps could involve exploring feature scaling, trying different classification algorithms, or performing hyperparameter tuning to potentially improve model performance.
